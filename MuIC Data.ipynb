```python
from ROOT import TH1F, TLegend, TObject, TPaveText
import numpy as np
from ROOT import TCanvas

line_id = 0

eta_scatter_hist =TH1F ('eta scattering', 'GEN #Delta #eta', 100,-0.004,0.004)

for event in root_file.events:
    for gen_id in range (event. vsize_gen):
        eta_scatter = event.gen_eta_scatter[gen_id]
        
        eta_scatter_hist.Fill(eta_scatter)
        line_id +=1
canvas = TCanvas ('eta scatter', 'GEN #eta')
canvas.cd(0)

eta_scatter_hist.Fit ("gaus","","")

Const = eta_scatter_hist.GetFunction("gaus").GetParameter(0);
Mean = eta_scatter_hist.GetFunction("gaus").GetParameter(1);
Sigma = eta_scatter_hist.GetFunction("gaus").GetParameter(2);

Const= round(Const,2)
Mean = round(Mean, 2)
Sigma = round (Sigma, 2)


legend = TPaveText(-1, 500, -3, 580)
legend.AddText(f"Const:{Const}")
legend.AddText(f"Mean: {Mean}")
legend.AddText(f"Standard dev: {Sigma}")

eta_scatter_hist.Draw('')
canvas.Draw('')
legend.Draw('')
```

```python
from ROOT import TH1F, TLegend, TObject, TPaveText
import numpy as np
from ROOT import TCanvas

line_id = 0


energy_deposit_hist =TH1F ('energy_deposited', 'GEN #Delta E',100, 0,beam_energy)

for event in root_file.events:
    for gen_id in range (event.vsize_gen):
        gen_e_deposited_nozzle = event.gen_e_deposited_nozzle[gen_id]
        
        energy_deposit_hist.Fill(gen_e_deposited_nozzle)
             
        line_id +=1
        

canvas =TCanvas ('gen_e_energy_deposited', 'GEN #Delta E')
canvas.cd(0)
energy_deposit_hist.Fit("gaus", "","",)
energy_deposit_hist.GetXaxis().SetTitle("E")

Const = energy_deposit_hist.GetFunction("gaus").GetParameter(0);
Mean = energy_deposit_hist.GetFunction("gaus").GetParameter(1);
Sigma = energy_deposit_hist.GetFunction("gaus").GetParameter(2);

Const= round(Const,3)
Mean = round(Mean, 3)
Sigma = round (Sigma,3)



legend = TPaveText(1400, 40, 1800, 60)
legend.AddText(f"Const:{Const}")
legend.AddText(f"Mean: {Mean}")
legend.AddText(f"Standard dev: {Sigma}")


energy_deposit_hist.Draw('')
legend.Draw('')
canvas.Draw('')
    
```
